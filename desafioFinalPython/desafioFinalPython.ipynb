Questão:
a.Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department);

```python
df2 = df.groupby('Department')[['EmployeeCount']].count()
df2.plot(kind='bar')
df.groupby('Department')[['EmployeeCount']].count()
```

Insight:

* A área de pesquisa e desenvolvimento é a que possui o maior número de funcionários.
* O setor de recursos humanos é o que tem menos pessoas.





Questão:
b.Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education);

```python
df2 = df.groupby('Education')[['Education']].count()
df2.plot(kind='line')
df.groupby('Education')[['Education']].count()
```

Insight:

* A maioria dos funcionários possui nível de escolaridade 3.0.
* Poucos funcionários alcançam o nível de escolaridade 5.0

Questão:
c.Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition).



```python
df2 = df.groupby('Attrition')[['Attrition']].count()
df2.plot(kind='pie', y='Attrition', autopct='%1.1f%%',)
df.groupby('Attrition')[['Attrition']].count()
```

Insight:
* Mais de 80% não tem atrito com a empresa
* Mais de 15% tem atrito com a empresa
